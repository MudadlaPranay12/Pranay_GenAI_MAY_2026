```python
import google.generativeai as genai
import time
from google.colab import userdata


genai.configure(api_key=userdata.get('GEMINI_API_KEYS'))
```

```python

knowledge_base = """
SYSTEM: Smart Road Emergency Assistant

==== OVERVIEW ====
- Q: What is the Smart Road Emergency System?
  A: It is an AI + IoT based system that helps ambulances get priority on roads by clearing traffic automatically and guiding drivers.

- Q: What is the main goal of this system?
  A: The main goal is to reduce emergency response time and help patients reach hospitals faster.

==== AMBULANCE PRIORITY SYSTEM ====
- Q: How does ambulance get priority on roads?
  A: Traffic signals automatically turn green in the direction of the ambulance using smart signal control.

- Q: What happens when an ambulance is detected?
  A: Nearby traffic signals adjust and vehicles are guided to clear the emergency lane.

- Q: Can multiple ambulances get priority?
  A: Yes, the system prioritizes based on proximity and urgency to manage multiple emergency vehicles.

==== DRIVER ALERT SYSTEM ====
- Q: How are drivers alerted?
  A: Drivers receive mobile notifications, sound alerts, and visual warnings to clear the road.

- Q: What should drivers do when they receive an alert?
  A: Drivers should move to the left side and immediately clear the lane for the ambulance.

- Q: What if a driver ignores the alert?
  A: The system may log the violation and notify traffic authorities.

==== GPS & ROUTING ====
- Q: How does GPS help in this system?
  A: GPS tracks the ambulance in real time and helps find the fastest route based on traffic conditions.

- Q: Can the ambulance route change during travel?
  A: Yes, routes are updated dynamically based on real-time traffic.

- Q: What happens if GPS signal is weak?
  A: The system uses last known location and continues updating when signal is restored.

==== TRAFFIC SIGNAL CONTROL ====
- Q: How do traffic signals know an ambulance is coming?
  A: Signals receive real-time data from the ambulance tracking system.

- Q: Can traffic signals be controlled manually?
  A: Yes, manual override is possible in case of emergencies or system failure.

- Q: Does the system work at all junctions?
  A: It works at smart-enabled junctions equipped with IoT traffic controllers.

==== EMERGENCY EDGE CASES ====
- Q: What happens if two ambulances come from different directions?
  A: The system prioritizes based on urgency and shortest delay impact.

- Q: What if traffic is very heavy?
  A: The system dynamically reroutes vehicles and adjusts signals to reduce congestion.

- Q: Does the system prevent all accidents?
  A: No, but it significantly reduces emergency response delays and improves safety.

==== GENERAL ====
- Q: Is this system real or simulation?
  A: It is a simulation model of a smart traffic management system using AI and IoT concepts.

- Q: Where can this system be used?
  A: It can be used in smart cities, highways, and urban traffic control systems.

- Q: Does it work without internet?
  A: Basic functions may work locally, but real-time coordination requires network connectivity.
"""

# ================= SYSTEM INSTRUCTION =================

system_instruction = f"""
You are a Smart Road Emergency Assistant chatbot.

RULES:
1. Answer ONLY using the knowledge base below.
2. If question is not present, say:
   "Sorry, I don't have that information. Please contact support team."
3. Be short, clear, and helpful (1-3 lines max).
4. Do not invent answers.
5. Be friendly and professional.

=== KNOWLEDGE BASE ===
{knowledge_base}
"""

model = genai.GenerativeModel(
    model_name='gemini-2.5-flash-lite',
    system_instruction=system_instruction
)

chat = model.start_chat(history=[])

print("Smart Road Emergency Assistant")
print("Type 'exit' to stop\n")

while True:
    user_input = input("You: ")

    if user_input.lower() in ["quit", "bye", "exit"]:
        print("Bot: Goodbye! Stay safe 🚦")
        break

    reply = chat.send_message(
        user_input,
        generation_config={
            "temperature": 0.2,
            "max_output_tokens": 200
        }
    )

    print(f"Bot: {reply.text}\n")
    time.sleep(2)
```
