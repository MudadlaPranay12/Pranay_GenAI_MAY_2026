```python
# Part A

def add_item(item, cart=[]):
    cart.append(item)
    return cart


print(add_item("apple"))
print(add_item("banana"))
print(add_item("milk", cart=["bread"]))
print(add_item("eggs"))

#Error : cart=[]
#This list is created once when function is defined, not every time function runs.
#So Python keeps reusing same list.
```

```python
def add_item(item, cart=None):
    if cart is None:
        cart = []

    cart.append(item)
    return cart


print(add_item("apple"))
print(add_item("banana"))
print(add_item("milk", cart=["bread"]))
print(add_item("eggs"))
```

```python
# Part C

def create_cart(owner, discount=0):
    return {
        "owner": owner,
        "items": [],
        "discount": discount
    }


def add_to_cart(cart, name, price, qty=1):
    cart["items"].append({
        "name": name,
        "price": price,
        "qty": qty
    })


def update_price(price_tuple, new_price):
    price_tuple[0] = new_price


def calculate_total(cart):
    total = 0

    for item in cart["items"]:
        total += item["price"] * item["qty"]

    discount_amount = total * cart["discount"] / 100
    final_total = total - discount_amount

    return final_total


cart1 = create_cart("Aarav", 10)
cart2 = create_cart("Ravi", 5)

add_to_cart(cart1, "Apple", 50, 2)
add_to_cart(cart1, "Milk", 30, 1)

add_to_cart(cart2, "Bread", 20, 3)
add_to_cart(cart2, "Eggs", 10, 2)

print("\nCart 1:")
print(cart1)

print("Cart 1 Total:", calculate_total(cart1))

print("\nCart 2:")
print(cart2)

print("Cart 2 Total:", calculate_total(cart2))


price_tuple = (100,)

try:
    update_price(price_tuple, 200)
except TypeError:
    print("\nTuples are immutable")
```

```python
# Discussion Answers

# Why is discount=0 safe but cart=[] dangerous?
# discount=0 is immutable.
# cart=[] is mutable and same list gets reused.

# Difference between rebinding and mutating:
# Mutating = changing same object
# Rebinding = assigning new object

# Mutable:
# list, dict, set

# Immutable:
# tuple, str, int

# When list is passed into function and modified,
# changes reflect outside because same object is modified.
```
