#Blog generator


```python
import google.generativeai as genai
genai.configure(api_key="AIzaSyBuYXYW4cVX31USC4jbVubk9yvehHoTrO8 ")
```

```python
def generate_blog(topic, tone, length_words=300):

    system_prompt = (
        "You are a professional blog writer. "
        "You write engaging, informative, and easy-to-read blogs."
    )

    user_prompt = f"""
    Write a blog post.

    Topic: {topic}
    Tone: {tone}
    Length: approximately {length_words} words
    """

    model = genai.GenerativeModel(
        model_name="gemini-2.5-flash-lite",
        system_instruction=system_prompt
    )


    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.7,
            "max_output_tokens": 700
        }
    )

    return response.text


```

```python

topic = input("Enter Blog Topic: ")
tone = input("Enter Blog Tone: ")
length = int(input("Enter Blog Length in Words: "))

print("\nGenerating Blog...\n")


blog = generate_blog(topic, tone, length)
print(blog)
```

#Email generator


```python
import google.generativeai as genai
genai.configure(api_key="AIzaSyBuYXYW4cVX31USC4jbVubk9yvehHoTrO8 ")
```

```python
def generate_email(recipient, purpose, tone):

    system_prompt = (
        "You are a professional email writer "
        "who writes clear and polite emails."
    )

    user_prompt = f"""
    Write a professional email.

    Recipient: {recipient}
    Purpose: {purpose}
    Tone: {tone}
    """

    model = genai.GenerativeModel(
        model_name='gemini-2.5-flash-lite',
        system_instruction=system_prompt
    )

    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.5,
            "max_output_tokens": 400
        }
    )

    return response.text
```

```python
recipient = input("Enter Recipient Name: ")
purpose = input("Enter Email Purpose: ")
tone = input("Enter Email Tone: ")

print("\nGenerating Email...\n")

print(generate_email(recipient, purpose, tone))
```
